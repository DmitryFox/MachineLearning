```python
import math
import random

import matplotlib.pyplot as plt
import numpy as np

from sklearn.metrics import mean_squared_error
from lasso import Lasso
```

```python
# Генерируем данные для тренировки
# Вопросы
x = np.arange(1, 10, 0.1)
# Оригинальные ответы
y = np.zeros(x.size)
# Ответы с помехами
noise_y = np.zeros(x.size)

for i in range(y.size):
    y[i] = x[i]
    # Оригинальные ответы + шум
    noise_y[i] = y[i] + (math.sin(x[i]) * 4) * random.random() + (random.random() * 15 if random.random() < 0.20 else 0.)
```

```python
plt.scatter(x, y, s=5)
plt.scatter(x, noise_y, s=5, color='green')

tX = np.array([x]).T

model = Lasso(alpha=0.1)
model.fit(tX, noise_y)

result = model.predict(tX)
plt.plot(sorted(x), result, 'b', label='LASSO', color='GREEN')

print("LASSO MSE %r" % mean_squared_error(y, result))

plt.legend(bbox_to_anchor=(1.017, -0.1))
plt.tight_layout(pad=1.7)
plt.title("LASSO")

plt.show()
```
